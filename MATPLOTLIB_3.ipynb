```python
1: Pie Chart with Custom Formatting
Task: Create a pie chart to show the market share of different smartphone brands. 
Use the following data: brands = ['Apple', 'Samsung', 'Xiaomi', 'Huawei', 'Others'] 
and shares = [30, 25, 20, 15, 10]. 
Customize the chart with percentage labels, a shadow effect, 
and an exploded slice for the largest share. Add a title and a legend.  
```

```python
import matplotlib.pyplot as plt
import numpy as np

define = ['Apple', 'Samsung', 'Xiaomi', 'Huawei', 'Others'] 
y = [30, 25, 20, 15, 10]
colors = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen', 'hotpink']
exp = (0.2, 0, 0, 0,0) 
plt.pie(y, labels = define, explode=exp, colors=colors,
        autopct='%1.1f%%', shadow=True) 
plt.title("Market share of different smartphone brands")
plt.legend()
plt.show()
```

```python
2: Box Plot for Data Comparison
Task: Create a box plot to compare the test scores of three different classes. 
Generate random data for each class using np.random.normal(): 
Class A (mean=70, std=10, n=50), Class B (mean=75, std=8, n=50), and Class C (mean=65, std=12, n=50). 
Customize the box plot with colors, add a title, and label the axes. 
Include a grid for better readability.
```

```python
import matplotlib.pyplot as plt
import numpy as np
class_A = np.random.normal(loc=70, scale=10, size=50)
class_B = np.random.normal(loc=75, scale=8, size=50)
class_C = np.random.normal(loc=65, scale=12, size=50)
boxprops = dict(facecolor='lightblue', color='blue')
medianprops = dict(color='red', linewidth=2)
flierprops = dict(marker='o', markerfacecolor='orange', markersize=6)
plt.boxplot([class_A, class_B, class_C],
            notch=True,
            patch_artist=True,
            labels=['Class A', 'Class B', 'Class C'],
            boxprops=boxprops,
            medianprops=medianprops,
            flierprops=flierprops)
plt.title('Comparison of Test Scores Across Classes')
plt.xlabel('Class')
plt.ylabel('Scores')
plt.grid(True)


plt.show()
```

```python
: Area Plot
Task: Create an area plot to visualize the cumulative revenue from three product lines over six months. 
Use the following data: months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'], 
product_a = [50, 60, 80, 100, 120, 150], 
product_b = [30, 40, 50, 60, 80, 100], 
and product_c = [20, 25, 30, 35, 40, 50]. 
Stack the areas and add transparency for better visibility. 
Include a title, labels, and a legend.                                                                    
```

```python
import matplotlib.pyplot as plt
import numpy as np
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
x = [50, 60, 80, 100, 120, 150], 
y = [30, 40, 50, 60, 80, 100],
z = [20, 25, 30, 35, 40, 50]
plt.stackplot(months, x, y, z, 
labels=['Product A', 'Product B', 'Product C'], alpha=0.8)
plt.title = ("Cumulative Revenue from Product Lines")
plt.xlabel = ('Month')
plt.ylabel = ('Revenue')
labels=['Product A', 'Product B', 'Product C']
plt.legend(loc='upper left')
plt.grid(True)
plt.legend()
plt.show()
```

```python
Contour Plot
Task: Create a contour plot to visualize a 3D function 
z=sin(x)+cos(y) 
z = sin(x) + cos(y) 
z=sin(x)+cos(y) 
over a grid where x x x and y y y range from -5 to 5. 
Use a meshgrid with 100 points in each direction, add a colorbar, and label the axes and title. 
Customize the contour levels and add labels to the contour lines.
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(X) + np.cos(Y)

contours = plt.contour(X, Y, Z, levels= 60, cmap='viridis') 
plt.colorbar(contours)
plt.title('Contour Plot of z = sin(x) + cos(y)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
