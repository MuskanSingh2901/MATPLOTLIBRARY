```python
1. Create a line plot to visualize the growth of a population over 10 years. 
Use the following data: years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019] 
and population = [1000, 1200, 1500, 1800, 2200, 2700, 3300, 4000, 4800, 5700]. 
Add labels for the axes and a title.    
```

```python
import matplotlib.pyplot as plt

x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019] 
y = [1000, 1200, 1500, 1800, 2200, 2700, 3300, 4000, 4800, 5700]
colors = ["red", "green", "blue", "orange", "purple", "pink", "brown", "black", "magenta", "cyan"]
plt.scatter(x, y, c=colors, s=100, marker='o') 
plt.plot(x, y, color='gray', linestyle='-.', alpha=0.6)
plt.title("Population Growth")
plt.xlabel("Years")
plt.ylabel("Population")

plt.show()
```

```python
2: Bar Chart
Task: Create a bar chart to display the sales of different products. 
Use the following data: products = ['Laptop', 'Phone', 'Tablet', 'Headphones'] and sales = [150, 200, 80, 120]. Customize the bars with different colors and add data labels above each bar.
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ['Laptop', 'Phone', 'Tablet', 'Headphones'] 
y = [150, 200, 80, 120]
colors = ["r", "g", "y", "b"]
plt.bar(x,y,alpha = 0.7, color = colors, edgecolor='black')
plt.title("SALES OF DIFFERENT PRODUCTS")
plt.xlabel("PRODUCTS")
plt.ylabel("SALES")

plt.show()
```

```python
3: Scatter Plot
Task: Create a scatter plot to visualize the relationship between hours studied 
and exam scores for a group of students. 
Use the following data: hours = [2, 4, 5, 4, 5, 3, 6, 7, 8, 2] 
and scores = [65, 75, 70, 80, 85, 70, 90, 95, 80, 60]. Add a trend line (linear fit) to the plot.                                                      
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = [2, 4, 5, 4, 5, 3, 6, 7, 8, 2] 
y = [65, 75, 70, 80, 85, 70, 90, 95, 80, 60]
colors = ["red", "green", "blue", "orange", "purple", "pink", "brown", "black", "magenta", "cyan"]
plt.scatter(x,y,alpha = 0.7, c = colors)
plt.title("HOURS STUDIED FOR AN EXAM")
plt.xlabel("HOURS")
plt.ylabel("SCORES")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = [2, 4, 5, 4, 5, 3, 6, 7, 8, 2] 
y = [65, 75, 70, 80, 85, 70, 90, 95, 80, 60]
colors = [20, 35, 45, 55, 65, 75, 85, 95, 105, 115]
plt.scatter(x,y,alpha = 0.8, c = colors, cmap = "inferno")
plt.title("HOURS STUDIED FOR AN EXAM")
plt.xlabel("HOURS")
plt.ylabel("SCORES")
plt.colorbar()
plt.show()
```

```python
4: Histogram
Task: Create a histogram to show the distribution of test scores for a class. 
Generate 100 random scores between 0 and 100 
using np.random.normal(75, 10, 100) to simulate a normal distribution 
with a mean of 75 and a standard deviation of 10. Use 10 bins 
and customize the histograms appearance.                  
```

```python
import matplotlib.pyplot as plt
import numpy as np


scores = np.random.normal(75, 10, 100)
plt.hist(scores, bins=10, color='hotpink', edgecolor='black', alpha=1)
plt.title("Distribution of Test Scores")
plt.xlabel("Score Ranges")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle=':', alpha=0.6)


plt.show()
```
